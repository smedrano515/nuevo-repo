```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
#width = 13.5           # ancho transversal de socavación (m)
#W = [1, 2, 50, 100]  # longitudes de socavación (m)
x = np.linspace(0, 1, 500)   # coordenada adimensional

plt.figure(figsize=(8, 6))
for width in range (14):
    L = width*8
    # y = W * 4 * (x/L)*(1 - x/L) → parametrizamos en x/L
    y = width * 4 * x * (1 - x)
    plt.plot(x * L, y, label=f'L = {L} m')

plt.title('Familia de parábolas de socavación (vista en planta)')
plt.xlabel('Longitud de socavación (m)')
plt.ylabel('Ancho de socavación (m)')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
